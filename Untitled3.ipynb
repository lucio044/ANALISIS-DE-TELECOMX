# ANALISIS TELECOM X

###Diccionario de datos

* `customerID`: número de identificación único de cada cliente.

* `Churn`: si el cliente dejó o no la empresa
* `gender`: género (masculino y femenino)
* `SeniorCitizen`: información sobre si un cliente tiene o no una edad igual o mayor a 65 años
* `Partner`: si el cliente tiene o no una pareja
* `Dependents`: si el cliente tiene o no dependientes
* `tenure`: meses de contrato del cliente
* `PhoneService`: suscripción al servicio telefónico
* `MultipleLines`: suscripción a más de una línea telefónica
* `InternetService`: suscripción a un proveedor de internet
* `OnlineSecurity`: suscripción adicional de seguridad en línea
* `OnlineBackup`: suscripción adicional de respaldo en línea
* `DeviceProtection`: suscripción adicional de protección del dispositivo
* `TechSupport`: suscripción adicional de soporte técnico, menor tiempo de espera
* `StreamingTV`: suscripción de televisión por cable
* `StreamingMovies`: suscripción de streaming de películas
* `Contract`: tipo de contrato
* `PaperlessBilling`: si el cliente prefiere recibir la factura en línea
* `PaymentMethod`: forma de pago
* `Charges.Monthly`: total de todos los servicios del cliente por mes
* `Charges.Total`: total gastado por el cliente

 ## Extracción 📌

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Abrir y cargar el archivo JSON
with open('TelecomX_Data.json') as f:
    data = json.load(f)

# 2. "Aplanar" los datos y crear el DataFrame
# Ahora 'pd' ya funcionará correctamente
df = pd.json_normalize(data)

# 3. Ver el resultado
df.head()
```

## Transformación 🔧

```python
df = pd.json_normalize(data)
df.columns = [col.split('.')[-1] for col in df.columns]
df.head()
```

```python
df.info()
```

```python
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
df['Total'] = df['Total'].fillna(0)
df['Cuentas_Diarias'] = (df['Monthly'] / 30).round(2)
print("Columnas actualizadas:")
print(df[['customerID', 'Monthly', 'Total', 'Cuentas_Diarias']].head())
```

```python
columnas_esp = {
'gender': 'Genero',
'SeniorCitizen': 'Adulto_Mayor',
'Partner': 'En_Pareja',
'Dependents': 'Dependientes',
'tenure': 'Meses_Antiguedad',
'PhoneService': 'Servicio_Telefonico',
'Contract': 'Tipo_Contrato',
'Monthly': 'Cobro_Mensual',
'Total': 'Cobro_Total',
'Churn': 'Evasion'
}
df.rename(columns=columnas_esp, inplace=True)
```

```python
df['Genero'] = df['Genero'].replace({'Female': 'Femenino', 'Male': 'Masculino'})
df['Evasion'] = df['Evasion'].replace({'Yes': 'Si', 'No': 'No'})
```

```python
df['Evasion_Binaria'] = df['Evasion'].map({'Si': 1, 'No': 0})
```

```python
print("Tabla Estandarizada:")
print(df[['Genero', 'Tipo_Contrato', 'Evasion', 'Evasion_Binaria']].head())
```

```python
df['Evasion_Binaria'] = df['Evasion_Binaria'].fillna(0)
df['Evasion_Binaria'] = df['Evasion_Binaria'].astype(int)
```

```python
print("Evasión Binaria corregida:")
print(df[['customerID', 'Evasion', 'Evasion_Binaria']].head())
```

## Carga y análisis 📊

```python
columnas_numericas = ['Meses_Antiguedad', 'Cobro_Mensual', 'Cobro_Total', 'Cuentas_Diarias']
analisis = df[columnas_numericas].describe().T
analisis['mediana'] = df[columnas_numericas].median()
print("=== ANÁLISIS DESCRIPTIVO DE TELECOM X ===")
print(analisis[['mean', 'mediana', 'std', 'min', 'max']].rename(columns={
'mean': 'Media',
'mediana': 'Mediana',
'std': 'Desv_Estandar',
'min': 'Mínimo',
'max': 'Máximo'
}))
```

```python
df = df[df['Evasion'].isin(['Si', 'No'])].copy()

conteo_evasion = df['Evasion'].value_counts()
print(f"✅ Registros procesados con éxito: {len(df)}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
colores_dict = {'No': '#66b3ff', 'Si': '#ff9999'}
colores_lista = [colores_dict[label] for label in conteo_evasion.index]

ax1.pie(conteo_evasion, labels=conteo_evasion.index, autopct='%1.1f%%',
startangle=90, colors=colores_lista, explode=[0.05]*len(conteo_evasion), shadow=True)
ax1.set_title('Proporción Real de Evasión')

sns.barplot(x=conteo_evasion.index, y=conteo_evasion.values,
hue=conteo_evasion.index, palette=colores_dict,
legend=False, ax=ax2)
ax2.set_title('Cantidad Exacta de Clientes')
ax2.set_ylabel('Número de Clientes')

for i, v in enumerate(conteo_evasion.values):
    ax2.text(i, v + 50, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()






```

* 🧐 Interpretación Final de la Distribución

- Tasa de Evasión (26.5%): Es una cifra crítica. En el mundo de los negocios, perder a más de una cuarta parte de tus clientes (1,869 personas) significa que hay algo en la oferta o en el servicio que no está convenciendo a largo plazo.

- Base Fiel (73.5%): Estos son los 5,174 clientes que sostienen la operación de Telecom X.

```python
variables = ['Genero', 'Tipo_Contrato', 'PaymentMethod']
fig, axes = plt.subplots(1, 3, figsize=(22, 7))

for i, var in enumerate(variables):
    sns.countplot(data=df, x=var, hue='Evasion',
palette={'No': '#66b3ff', 'Si': '#ff9999'},
ax=axes[i])

plt.tight_layout()
plt.show()
```

La evasión de clientes (26.5%) no es aleatoria, sino que está fuertemente ligada a la falta de compromiso contractual y a los métodos de pago manuales. El perfil del cliente con mayor riesgo de fuga es aquel que tiene un contrato mes a mes y paga mediante cheque electrónico.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=df, x='Evasion', y='Meses_Antiguedad',
hue='Evasion', palette={'No': '#66b3ff', 'Si': '#ff9999'},
legend=False, ax=ax1)

ax1.set_title('Antigüedad: ¿Cuándo se están yendo?', fontsize=14, fontweight='bold')
ax1.set_xlabel('¿Hubo Evasión?')
ax1.set_ylabel('Meses de Antigüedad')

sns.boxplot(data=df, x='Evasion', y='Cobro_Mensual',
hue='Evasion', palette={'No': '#66b3ff', 'Si': '#ff9999'},
legend=False, ax=ax2)

ax2.set_title('Cobros: ¿El precio es un factor de fuga?', fontsize=14, fontweight='bold')
ax2.set_xlabel('¿Hubo Evasión?')
ax2.set_ylabel('Monto del Cobro Mensual ($)')

plt.tight_layout()
plt.show()
```

La fuga en la empresa tiene un perfil claro: clientes nuevos con facturas altas. Estamos fallando en retener a los clientes de mayor valor durante su primer año de vida. El precio elevado, sumado a una posible falta de beneficios percibidos al inicio, está impulsando la evasión hacia la competencia.

```python
df['Evasion_Numerica'] = df['Evasion'].map({'Si': 1, 'No': 0})

df_numerico = df.select_dtypes(include=['number'])

matriz = df_numerico.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(matriz, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.8)

plt.title('Mapa de Calor: Correlación de Variables', fontsize=16, fontweight='bold')
plt.show()
```

Blindar a los clientes nuevos: La antigüedad es el factor de protección número uno.

Monitorear los cobros altos: El precio tiene una relación directa con la fuga; se debe equilibrar la relación costo-beneficio para los clientes premium.

# Informe Final: Análisis de Evasión de Clientes (Churn) - Telecom X

## 1. Introducción

El objetivo de este análisis es identificar los factores críticos que influyen en la evasión de clientes (Churn) de la empresa Telecom X. La pérdida de clientes no solo afecta los ingresos directos, sino que incrementa los costos de adquisición de nuevos usuarios. A través de la Ciencia de Datos, buscamos entender por qué el 26.5% de la base de clientes decide abandonar el servicio y cómo podemos prevenirlo.


## 2. Limpieza y Tratamiento de Datos

Para garantizar la calidad de los hallazgos, se realizaron los siguientes procesos técnicos:

Importación y Estructura: Se cargó el dataset y se identificaron variables categóricas (como género y contrato) y numéricas (como cobros y antigüedad).

Tratamiento de Datos Nulos: Se detectaron valores faltantes en la columna Cobro_Total. Estos registros se eliminaron para evitar sesgos, dejando una base sólida de 7,043 registros.

Transformación de Variables: Se crearon columnas numéricas auxiliares (Evasion_Binaria) para permitir el cálculo de correlaciones matemáticas.

## 3. Análisis Exploratorio de Datos (EDA)

Mediante visualizaciones estratégicas, identificamos patrones de comportamiento claros:

Perfil Contractual: Los clientes con contratos mes a mes representan la gran mayoría de las bajas.

Comportamiento del Pago: El método de Cheque Electrónico presenta una tasa de evasión desproporcionadamente alta en comparación con los pagos automáticos.

Variables Numéricas: El análisis de cajas (Boxplots) reveló que la fuga se concentra en los clientes nuevos (menos de 10 meses de antigüedad) y en aquellos con cargos mensuales elevados (superiores a $80)

## 4. Conclusiones e Insights

Tras cruzar los datos demográficos, financieros y operativos, concluimos lo siguiente:

La Antigüedad es el mejor protector: Existe una correlación negativa de -0.35 entre el tiempo de permanencia y la evasión. A mayor tiempo en la empresa, menor es el riesgo de fuga.

Sensibilidad al Precio: Los clientes "Premium" (que pagan más) son los más propensos a irse, lo que indica que no perciben un valor proporcional al costo o son atraídos por ofertas competitivas.

Fricción en Pagos: Los métodos de pago manuales generan puntos de abandono, mientras que la automatización del cobro fomenta la retención.

## 5. Recomendaciones Estratégicas

Basándonos en los datos, sugerimos las siguientes acciones para reducir el Churn:

Plan de "Bienvenida Crítica": Implementar un programa de seguimiento intensivo durante los primeros 6 meses de vida del cliente, ofreciendo bonos de lealtad para superar la etapa de mayor riesgo.

Incentivo a la Automatización: Ofrecer descuentos o beneficios adicionales a los clientes que migren su pago de Cheque Electrónico a Cargo Automático.

Conversión de Contratos: Diseñar campañas para convertir a los usuarios de "mes a mes" hacia contratos anuales mediante planes de ahorro, "amarrando" la lealtad a largo plazo.
